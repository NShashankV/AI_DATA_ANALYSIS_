```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create DataFrame
data = pd.DataFrame({
    'Height_cm': heights,
    'Weight_kg': weights
})

# Create scatter plot
sns.scatterplot(x='Height_cm', y='Weight_kg', data=data, color='teal', s=100)

# Add title and labels
plt.title('Height vs. Weight Scatter Plot')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

plt.show()
```
