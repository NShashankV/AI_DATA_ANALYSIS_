```python
#
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
sizes = [10, 15, 13, 18, 12]  # Size for each point

# Create DataFrame
data = pd.DataFrame({
    'Height_cm': heights,
    'Weight_kg': weights,
    'Size': sizes
})

# Create scatter plot with size parameter
sns.scatterplot(x='Height_cm', y='Weight_kg', size='Size', data=data, legend='brief', sizes=(50, 300), color='purple')

# Add title and labels
plt.title('Height vs. Weight Scatter Plot with Point Sizes')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

plt.show()
```
