```python
#
import matplotlib.pyplot as plt

# Data for bar plot
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]  # in thousands

# Data for scatter plot
heights = [160, 170, 165, 180, 175]  # cm
weights = [55, 65, 60, 80, 76]       # kg

# Create subplots: 1 row, 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot on the first subplot
axs[0].bar(products, sales, color='skyblue')
axs[0].set_title('Sales of Products (in thousands)')
axs[0].set_xlabel('Products')
axs[0].set_ylabel('Sales (Thousands)')
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter plot on the second subplot
axs[1].scatter(heights, weights, color='green')
axs[1].set_title('Height vs. Weight Scatter Plot')
axs[1].set_xlabel('Height (cm)')
axs[1].set_ylabel('Weight (kg)')
axs[1].grid(True, linestyle='--', alpha=0.7)

# Adjust layout so titles and labels don’t overlap
plt.tight_layout()
plt.show()
```
