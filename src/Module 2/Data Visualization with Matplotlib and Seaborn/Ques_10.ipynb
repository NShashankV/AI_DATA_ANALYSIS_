```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset('tips')

# Display first few rows (optional)
print(tips.head())

# Create a scatter plot with multiple features
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',     # Color by smoker status
    style='sex',      # Different markers for gender
    size='size',      # Size based on party size
    sizes=(40, 200),  # Range of sizes
    palette='Set2'
)

# Add title and axis labels
plt.title('Total Bill vs. Tip with Smoker & Gender Info')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside
plt.tight_layout()
plt.show()
```
