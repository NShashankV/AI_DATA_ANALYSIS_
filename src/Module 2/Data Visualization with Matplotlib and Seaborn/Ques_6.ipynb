```python

import matplotlib.pyplot as plt

# Data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]  # in thousands

# Create bar plot with customizations
plt.bar(products, sales, color='#FF6347')  # Tomato red color

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Customize fonts for title and labels
plt.title('Sales of Products (in thousands)', fontsize=16, fontweight='bold', fontname='Comic Sans MS')
plt.xlabel('Products', fontsize=14, fontweight='bold', fontname='Arial')
plt.ylabel('Sales (Thousands)', fontsize=14, fontweight='bold', fontname='Arial')

plt.show()
```
