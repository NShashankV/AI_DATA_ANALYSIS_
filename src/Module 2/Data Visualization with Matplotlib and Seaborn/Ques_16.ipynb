```python

import matplotlib.pyplot as plt

# Sample data
products = ['A', 'B', 'C', 'D', 'E']
sales = [23, 75, 57, 90, 65]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
revenue = [120, 150, 180, 160, 200]

# Create a figure with 1 row and 2 columns of subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# First subplot: Bar plot
axes[0].bar(products, sales, color='skyblue')
axes[0].set_title('Product Sales')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Sales (in 1000s)')
axes[0].grid(True, linestyle='--', alpha=0.5)

# Second subplot: Line plot
axes[1].plot(months, revenue, marker='o', linestyle='-', color='teal')
axes[1].set_title('Monthly Revenue')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Revenue ($)')
axes[1].grid(True, linestyle='--', alpha=0.5)

# Adjust layout to avoid overlap
plt.tight_layout()
plt.suptitle('Combined Sales and Revenue Data', fontsize=16, y=1.05)
plt.show()
```
