```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Customer Dataset
# - Load a custom CSV file named customer_data.csv .
# - Find any missing values in specific columns like 'Email' and 'Phone'.






# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Duplicate Emails in Customer Dataset
# - Identify duplicate emails which might indicate duplicate customer records.



import pandas as pd

# Load the customer dataset
df = pd.read_csv('customer_data.csv')

# Print all columns to verify names
print("Columns in dataset:", df.columns.tolist())

# Define columns to check for missing values (update these if needed)
columns_to_check = ['Email', 'Phone']

# Filter only columns that exist in the dataset
existing_cols = [col for col in columns_to_check if col in df.columns]

if existing_cols:
    print(f"\nðŸ” Missing values in columns: {existing_cols}")
    print(df[existing_cols].isnull().sum())
else:
    print("\nâš ï¸ None of the specified columns (Email, Phone) found in the dataset.")

# ---------------------------------------------
# Identify Duplicates in Email column (if exists)
# ---------------------------------------------
if 'Email' in df.columns:
    duplicate_emails = df[df.duplicated(subset='Email', keep=False)]
    print("\nðŸ“§ Duplicate Email Records:")
    if not duplicate_emails.empty:
        print(duplicate_emails)
    else:
        print("No duplicate emails found.")
else:
    print("\nâš ï¸ 'Email' column not found; skipping duplicate email check.")

# ---------------------------------------------
# Generate Data Quality Report
# ---------------------------------------------

print("\nðŸ“‹ Data Quality Report:")

# Missing values per column
missing_report = df.isnull().sum()
print("\nðŸ” Missing values per column:")
print(missing_report[missing_report > 0])

# Duplicate emails count
if 'Email' in df.columns:
    duplicate_email_count = df.duplicated(subset='Email').sum()
    print(f"\nðŸ“§ Total duplicate emails: {duplicate_email_count}")

# Total duplicate rows
total_duplicate_rows = df.duplicated().sum()
print(f"ðŸ§¯ Total duplicate rows: {total_duplicate_rows}")

# Basic descriptive statistics
print("\nðŸ“Š Descriptive Statistics:")
print(df.describe(include='all'))

# Part 3: Generate a Data Quality Report

# Task 3: Customer Dataset Report
# - Summarize the data quality with missing values, duplicates, and inconsistencies for customer_data.csv .





```
